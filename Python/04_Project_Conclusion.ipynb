# Customer Churn Analysis - Project Conclusion

## Objective

The objective of this project was to analyze customer churn, identify the key factors influencing customer retention, build predictive machine learning models, and provide business recommendations to reduce customer churn.

## Key Findings

- Customers with month-to-month contracts showed the highest churn rate.
- Customers using fiber optic internet services were more likely to leave.
- Electronic check users exhibited higher churn than customers using automatic payment methods.
- Customers with higher monthly charges had a greater tendency to churn.
- Customers with longer tenure were significantly more likely to remain with the company.

## Machine Learning Summary

Two machine learning models were trained to predict customer churn:

- Logistic Regression
- Random Forest Classifier

The models were evaluated using:
- Accuracy
- Confusion Matrix
- Classification Report

The model with the higher accuracy was selected as the final prediction model.

## Business Recommendations

- Encourage customers to switch from month-to-month to long-term contracts through loyalty offers.
- Provide targeted discounts or personalized offers to customers with high monthly charges.
- Improve customer engagement during the initial months of service to reduce early churn.
- Promote automatic payment methods to improve customer retention.
- Continuously monitor high-risk customer segments using predictive analytics.

## Conclusion

This project demonstrates an end-to-end customer churn analysis workflow, including data cleaning, SQL-based business analysis, exploratory data analysis, feature engineering, predictive modeling, and business recommendation generation using SQL, Python, Machine Learning, and Power BI.
